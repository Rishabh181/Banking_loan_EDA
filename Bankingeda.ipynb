```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns' , None)
```

```python
df['Estimated Income'].min()
```

```python
bins = [0 , 100000 , 300000 , float('inf')]
labels = ['Low' , 'Med' , 'High']

df['Income Band'] = pd.cut(df['Estimated Income'] , bins = bins , labels = labels , right =False )
```

```python
df['Income Band'].value_counts().plot(kind= 'bar')
```

```python
# Examime the distribution of unique categories in categorical columns

categorical_cols = [feature for feature in df.columns if df[feature].dtype == 'object']
```

```python
categorical_cols
```

```python
categorical_cols.add('IAId','BRId' , 'GenderId' , 'Amount of Credit Cards', 'Properties Owned', 'Risk Weighting')
```

```python
categorical_cols = df[['IAId','BRId' , 'GenderId' , 'Amount of Credit Cards','Nationality','Occupation', 'Fee Structure' ,'Loyalty Classification' , 'Properties Owned', 'Risk Weighting', 'Income Band']]
categorical_cols
```

```python
# Univariate analysis

for col in categorical_cols:
    print(f"values count for '{col}' :")
    display(df[col].value_counts())
```

```python
# Bivariate analysis

for i , predictor in enumerate(df[['IAId','BRId' , 'GenderId' , 'Amount of Credit Cards','Nationality','Occupation', 'Fee Structure' ,'Loyalty Classification' , 'Properties Owned', 'Risk Weighting', 'Income Band']]) :
    plt.figure(i)
    sns.countplot(data = df , x= predictor , hue = 'GenderId')
```

```python
for i , predictor in enumerate(df[['IAId','BRId' , 'GenderId' , 'Amount of Credit Cards','Nationality','Occupation', 'Fee Structure' ,'Loyalty Classification' , 'Properties Owned', 'Risk Weighting', 'Income Band']]) :
    plt.figure(i)
    sns.countplot(data = df , x= predictor , hue = 'Nationality')
```

```python
# HISTplot of value count for different Occupation

for col in categorical_cols:
    if col == "Occupation" :
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# Numerical Analysis

```python
df
```

```python
Numerical_cols = ['Estimated Income' , 'Superannuation Savings' , 'Credit Card Balance','Bank Loans' , 'Bank Deposits', 'Checking Accounts' , 'Saving Accounts' , 'Foreign Currency Account' , 'Business Lending']

# Univarite analysis and visualization
plt.figure(figsize= (20,15))
for i, col in enumerate(Numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col] , kde = True)
    plt.title(col)
plt.show()
```

```python
# all of them are skewed
Numerical_cols = ['Estimated Income' , 'Superannuation Savings' , 'Credit Card Balance','Bank Loans' , 'Bank Deposits', 'Checking Accounts' , 'Saving Accounts' , 'Foreign Currency Account' , 'Business Lending']

correlation_matrix = df[Numerical_cols].corr()

plt.figure(figsize=(12,12))

sns.heatmap(correlation_matrix , annot = True , cmap = 'crest' , fmt =".2f")

plt.show()
```
